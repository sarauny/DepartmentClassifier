## Pre-processing

Before classify each job title into departments, following steps are conducted as pre-processing to achieve higher matching rate between words used in job titlles and keywords of departments.
- Normalizing
    1. Replace non-alphabeltical characters with a single spacing
    2. lowercasing to all single words
    3. Tokenizing all sentences to single words in list
    4. Removing stopwords
- Stemming words
    - To have the maximum sensivity of departmental assignment based on the keywords, an aggresive stemming method, ie. LancasterStemmer was used


```python
# adding additional meaningless stopwords for the dataset (based on authors' perspective on her current approach)
stopwords_extended = stopwords.words("english") + \
    ['senior', 'sr', 'snr', 'ceo', 'manager', 'deputy', 'associate', 
     'director', 'vice', 'asst', 'assistant', 'president', 'supervisor']
```

```python
# normalizing and stemming the department keywords  
stemDeptKeywords = copy.deepcopy(data)
        
for datum in stemDeptKeywords:
    for key in datum:
        datum[key] = list(traverse(datum[key]))
        datum[key] = [re.sub(r'[^a-zA-Z0-9]', ' ', string) for string in datum[key]]
        datum[key] = [word_tokenize(word.lower()) for word in datum[key]]
        datum[key] = list(traverse(datum[key]))
        datum[key] = [word for word in datum[key] if word not in stopwords_extended] 
        datum[key] = [LancasterStemmer().stem(word) for word in datum[key]]
```

```python
# normalizing and stemming the job title keywords  
#df = copy.deepcopy(df[:1000])
df['JobTitleKeys'] = df['Job Title']

def tokenize_column(df):
    for i in range(len(df)):
        try:
            df['JobTitleKeys'].iloc[i] = re.sub(r'[^a-zA-Z0-9]', ' ', df['Job Title'].iloc[i])
            df['JobTitleKeys'].iloc[i] = word_tokenize(df['JobTitleKeys'].iloc[i].lower())
            df['JobTitleKeys'].iloc[i] = list(traverse(df['JobTitleKeys'].iloc[i]))
            df['JobTitleKeys'].iloc[i] = [word for word in df['JobTitleKeys'].iloc[i] if word not in stopwords_extended]
            df['JobTitleKeys'].iloc[i] = [LancasterStemmer().stem(word) for word in df['JobTitleKeys'].iloc[i]]
        except:
            None 
    return df
    
df_processed = pd.DataFrame()
df_processed = process_Pandas_data(tokenize_column, df, None, 4)
print("The shape of the df_processed is: ", df_processed.shape)
df_processed[:5]
```

## A simple cascading method to approach the problem

### 1. Initial simple keywords matchings for departmental assignment

Simple keyword matching approach with some NLTK tools seems to work the best because of the small number of words within the job title data (Author's perspective). I have decided to simply match each normalized and stemmed words within job titles with ones from the departmental job scopes. 

 For example, a job title called "senior manager operations" becomes "op" after pre-processing steps above. Since "op" could be found on the table of keywords under "Management" department, this job title belongs to "Management". 

```python
# matching department name according to job titles
df_processed['Department'] = df_processed.apply(lambda x: [], axis=1) 

# defining keywords matching functions for Departmental Data
def match_keywords(df_chunks):
    for i in range(len(df_chunks)):
        for word in df_chunks['JobTitleKeys'].iloc[i]:
            for datum in stemDeptKeywords:
                for key in datum:
                    if word in datum[key]:
                        df_chunks['Department'].iloc[i].append(str(key))
                    
    return df_chunks

df_processed_matched = pd.DataFrame()
df_processed_matched = process_Pandas_data(match_keywords, df_processed, stemDeptKeywords)
df_processed_matched['Department'] = df_processed_matched['Department'].apply(set)
df_processed_matched[:5]
```

```python
print('After matching the JobTitleKeys with the table of keywords, we still have',\
      len(df_processed_matched[df_processed_matched['Department'] == set()]), '(' + \
      "{0:.2f}".format(len(df_processed_matched[df_processed_matched['Department'] == set()])*100/84577) + \
      '%)', 'job titles do not belong to any department.')
```

### 2. Rematching for the job titles that do not belong to any department
To achieve higher sensitivity, a new keywords table is created based on the existing matching result. Each keyword in a job title was assigned with all departments previously assigned to the job title.

For example, a job title called "senior analyst" becomes "analyst" after the previous pre-processing steps. However, analyst itself doesn't allow assignment of any department, although his/her job scope can be quite general and able to work in many departments. For instance, job titles as "financial analyst" and "operation analyst" may get assigned with "Finance" and "Management" departments, respectively. By leveraging the existing departmental assignment (previous Step 1 output as current input - a cascading approach), the new keywords table combines both "Finance" and "Management" departments for the job title keyword of "analyst". 

Please note that to avoid unspecific departmental assignment to unrelevant job titles, only those job titles without any departmental assignment from Step 1 and those job titles with only single keyword left after the pre-processing steps are further processed with following Step 2.

```python
def rematch_keywords(df_chunks):
    index=0
    df_chunks_JTT = pd.DataFrame(columns=['JobTitleKeys2', 'Department2', 'DepartmentUnion'])
    for i in range(len(df_chunks['JobTitleKeys'])):
        for j in range(len(df_chunks['JobTitleKeys'].iloc[i])):
            df_chunks_JTT.loc[index, 'JobTitleKeys2'] = df_chunks['JobTitleKeys'].iloc[i][j]
            df_chunks_JTT['Department2'].iloc[index] = df_chunks['Department'].iloc[i]
            index += 1
                    
    return df_chunks_JTT


JobTitleTable = pd.DataFrame()
JobTitleTable = process_Pandas_data(rematch_keywords, df_processed_matched)
JobTitleTable.reset_index(drop=True, inplace=True)

JobTitleTable.shape
```

```python
# group by JobTitleKeys2
GroupedKeys = JobTitleTable.groupby("JobTitleKeys2").apply(lambda x: set.union(*x.Department2))
```

```python
# when Department column is empty and there is only one word in JobTitleKeys column, fill department column referring JobTitleTable 
for i in range(len(df_processed_matched)):
    if df_processed_matched['Department'].iloc[i] == set():
        if len(df_processed_matched['JobTitleKeys'].iloc[i]) == 1:
            df_processed_matched['Department'].iloc[i] = \
            list(GroupedKeys[GroupedKeys.index==df_processed_matched['JobTitleKeys'].iloc[i][0]])

df_processed_matched[df_processed_matched['Department'] == set()]
df_processed_matched[:10]
```

```python
print('After rematching with the new table of keywords, we now have',\
      len(df_processed_matched[df_processed_matched['Department'] == set()]), '(' + \
      "{0:.2f}".format(len(df_processed_matched[df_processed_matched['Department'] == set()])*100/84577) + \
      '%)', 'job titles do not belong to any department.')
```

```python
# Randomly sampling 100 samples for evaluation
df_processed_matched_100samples = df_processed_matched.sample(n=100)
df_processed_matched_100samples.to_excel('Results_randomSampling100.xls')
```
